```python
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization
import re

#Функція розбиття тексту на речення
def split_into_sentences(text):
    sentences = re.split(r'[.!?]+', text)
    return [s.strip() for s in sentences if s.strip()]

with open("/content/lab1-main/eng texts.txt", "r", encoding="latin-1") as f:
    eng_texts = [line.strip() for line in f.readlines() if line.strip()]

with open("/content/lab1-main/ukr texts.txt", "r", encoding="mac_cyrillic", errors="replace") as f:
    ukr_texts = [line.strip() for line in f.readlines() if line.strip()]

#Розбиття текстів на речення
eng_sentences = [split_into_sentences(text) for text in eng_texts]
ukr_sentences = [split_into_sentences(text) for text in ukr_texts]

print("English sentences:")
for i, sents in enumerate(eng_sentences, 1):
    print(f"Text {i}:", sents)

print("\nUkrainian sentences:")
for i, sents in enumerate(ukr_sentences, 1):
    print(f"Text {i}:", sents)


english_flat = [sent for text in eng_sentences for sent in text]


vectorizer_int = TextVectorization(
    max_tokens=10000,
    output_mode='int',
    output_sequence_length=50
)
vectorizer_int.adapt(english_flat)
int_seq = vectorizer_int(english_flat).numpy()
print("\nINT sequences:")
print(int_seq)
print("Розмірність:", int_seq.shape)


vectorizer_binary = TextVectorization(
    max_tokens=10000,
    output_mode='binary'
)
vectorizer_binary.adapt(english_flat)
multi_hot = vectorizer_binary(english_flat).numpy()
print("\nMULTI-HOT vectors:")
print(multi_hot)
print("Розмірність:", multi_hot.shape)


vectorizer_count = TextVectorization(
    max_tokens=10000,
    output_mode='count'
)
vectorizer_count.adapt(english_flat)
count_vec = vectorizer_count(english_flat).numpy()
print("\nCOUNT vectors:")
print(count_vec)
print("Розмірність:", count_vec.shape)

#TF-IDF vectors
vectorizer_tfidf = TextVectorization(
    max_tokens=10000,
    output_mode='tf-idf'
)
vectorizer_tfidf.adapt(english_flat)
tfidf_vec = vectorizer_tfidf(english_flat).numpy()
print("\nTF-IDF vectors:")
print(tfidf_vec)
print("Розмірність:", tfidf_vec.shape)
```
